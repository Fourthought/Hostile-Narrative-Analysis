```python
sentences = [
    "We condem the Taliban Regime",
    "Afghanistan's people have been brutalized",
    "This is the world's fight",
    "This is the fight of all who believe in progress and pluralism, tolerance and freedom",
    "The enemy of America is not our many Muslim friends; it is not our many Arab friends.",
    "Our enemy is a radical network of terrorists, and every government that supports them",
    "We will starve terrorists of funding",
    "And on behalf of the American people, I thank the world for its outpouring of support",
    "They are some of the murderers indicted for bombing American embassies in Tanzania and Kenya, and responsible for bombing the USS Cole",
    "On September the 11th, enemies of freedom committed an act of war against our country."
]

for sentence in sentences:
    doc = nlp(sentence)
    matches = deps(doc)
    if matches:
        displacy.render(doc)
        print(sentence)
        display(pd.DataFrame(matches.get_repr()))
        
```

```python
filename = "20010920-Address to Joint Session of Congress Following 911 Attacks.txt"
bush = find_file(filename)
doc = nlp(bush)
for sent in doc.sents:
    matches = deps(nlp(sent.text))
    if matches:
        displacy.render(sent)
        print(sent)
        display(pd.DataFrame(matches))
```
