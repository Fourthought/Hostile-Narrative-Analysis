```python
import coreferee, spacy
nlp = spacy.load('en_core_web_md')
nlp.add_pipe("merge_entities")
nlp.add_pipe('coreferee')
```

```python
doc = nlp(text)

from spacy.tokens import Span, Token
Token.set_extension("coref", default = None, force = True)

def get_coref_string(doclike):
    
    strings = []
    
    for token in doclike:
        
        roots = token._.coref_chains.resolve(token)
        
        if roots:
            
            for root in roots:
                span = Span(root.doc, root.left_edge.i, root.right_edge.i + 1, label = root.ent_type_)
            strings.append(str(span))
        else:
            strings.append(token.text)
            
    return ' '.join(strings)

cache = []
correct = 0
incorrect = 0

for chunk in doc:
    chain_roots = chunk._.coref_chains.resolve(chunk)

    if chain_roots:
        print(chunk, '=>', chain_roots)
        print(chunk.sent)
        print(get_coref_string(chunk.sent))
        
        answer = input("correct (y) / incorrect (n)")
        if answer.lower() == 'y':
            correct += 1
        if answer.lower() == 'n':
            incorrect +=1
        print('-----')
        
accuracy = (correct / (correct + incorrect))* 100
```

```python
print(f"accuracy: {accuracy}")
print(f"correct: {correct}")
print(f"incorrect: {incorrect}")
```
