```python
import spacy
from spacy import displacy
import pandas as pd
from cndlib.cndutils import find_file
pd.set_option('display.max_colwidth',3000)

nlp = spacy.load("en_core_web_trf")
```

```python
text = """The American people are the ones who pay the taxes which fund the planes that bomb us in Afghanistan, \
the tanks that strike and destroy our homes in Palestine, the armies which occupy our lands in the Arabian Gulf, \
and the feats which ensure the blockade of Iraq."""
doc = nlp(text)
displacy.render(doc)
```

```python
import spacy
from spacy.tokens import Token
nlp = spacy.load("en_core_web_sm")
text = "shot shot"
predicate = nlp(text)[0]
predicate2 = nlp(text)[1]
print(f'{predicate=} ({predicate.i}), {predicate2=} ({predicate2.i})')

class Clause(dict):
    
    def is_linked(self, item):
        return item.get("predicate", False) == self.get("predicate", False)

clause = Clause(predicate=predicate)
clause2 = Clause(predicate=predicate2)
print(clause.is_linked(clause2))
```
